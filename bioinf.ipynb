```python
import sys
from Bio import AlignIO
from Bio.Align import AlignInfo
import numpy
```

```python
# to find the consensus
# read the file 
f = open("data/delta.fasta", "r")
deflines=[]   # names of sequences
sequences=[]   # data of sequences
morelines = True
counter=0   

while morelines:
    seq=""   
    moreseq = True
    while moreseq:
        nxtline = f.readline()
        if not nxtline:
            morelines = False
            break
        # if the first letter in the next line is ">" skip this line , 
        elif nxtline[0] != ">":
            seq+=nxtline.strip()
        else: 
            deflines.append(nxtline.strip())
            moreseq = False
    if counter!=0:
        sequences.append(seq)
    else:
        counter+=1
```

```python
# maximum length of sequences
length=min(len(sequences[0]) ,len(sequences[1]),len(sequences[2]),len(sequences[3]) ,len(sequences[4]),len(sequences[5]) ,len(sequences[6]),len(sequences[7]),len(sequences[8]) ,len(sequences[9]))
seq=""
seqlt=['A' ,'C' ,'G','T']

for i in range(length):
    arr=[0,0,0,0]
    arr=numpy.array(arr)
    for j in range(len(sequences)):
        lt=sequences[j][i]

        if lt=='A':
            arr[0]+=1
        elif lt=='C':
            arr[1]+=1
        elif lt=='G':
            arr[2]+=1
        elif lt=='T':
            arr[3]+=1

    max_index = arr.argmax()
    seq+=seqlt[max_index]

print(seq)
```

```python
from Bio import  SeqIO
fasta_sequnece=SeqIO.parse(open("data/data_dissmilarty.fasta"),'fasta')
seqs=[]
similarities=0
dissimilarities=0
for fasta in fasta_sequnece:
    seqs.append(str(fasta.seq))

maximamlength=max(len(seqs[0]) , len(seqs[1]),len(seqs[2]), len(seqs[3]),len(seqs[4]), len(seqs[5]),len(seqs[6]), len(seqs[7]), len(seqs[8]),len(seqs[9]),len(seqs[10]))
minlength=min(len(seqs[0]) , len(seqs[1]),len(seqs[2]), len(seqs[3]),len(seqs[4]), len(seqs[5]),len(seqs[6]), len(seqs[7]), len(seqs[8]),len(seqs[9]),len(seqs[10]))
for i in range(maximamlength):
    try:
        if i<minlength :
            if seqs[0][i]==seqs[1][i]==seqs[2][i]==seqs[3][i]== seqs[4][i]==seqs[5][i]==seqs[6][i]==seqs[7][i]==seqs[8][i]==seqs[9][i]==seqs[10][i]:
                similarities+=1
            else :
                dissimilarities+=1
        else:
            dissimilarities+=1

    except:
        pass

print("the maximum length:{}".format(maximamlength))
print("the minimum length:{}".format(minlength))
print("similarity:{}".format(similarities))
print("dissimilarity:{}".format(dissimilarities))
```
